```r
Convert("/newdatb/home/wanglab13/scratch/wholebrain/results/agg_HE-5-4W/scanpy/write/adata1_w5_1.h5ad",dest="/newdatb/home/wanglab13/scratch/wholebrain/results/agg_HE-5-4W/scanpy/write/adata1_w5_1_v1.h5Seurat", overwrite = TRUE)
```

```r
W5_3_RNA_neuron<- LoadH5Seurat(file = '/newdatb/home/wanglab13/scratch/wholebrain/results/agg_HE-5-4W/scanpy/write/adata1_w5_1_v1.h5Seurat')
metadata<-read.csv("/newdatb/home/wanglab13/scratch/wholebrain/results/agg_HE-5-4W/scanpy/write/adata1_w5_1_meatadata.csv")
metadata[,'Cells']<-names(W5_3_RNA_neuron@active.ident)
row.names(metadata)=metadata[,'Cells']
W5_3_RNA_neuron@meta.data=metadata
Idents(W5_3_RNA_neuron)<-'louvain_lineage'
```

```r
GW7_2 <- GW7
```

```r
W5_3_RNA_neuron_RG=subset(W5_3_RNA_neuron,idents = c('Myocyte_lineage','Skeletal_system_lineage','Intermediate_mesod_lineage','Cardiac_mesoderm_lineage','CNS_lineage','PNS_lineage'))
# W5_3_RNA_neuron_RG=subset(W5_3_RNA_neuron,idents = c('Myocyte_lineage','Skeletal_system_lineage','Intermediate_mesod_lineage','Cardiac_mesoderm_lineage','CNS_lineage'))
# W5_3_RNA_neuron_RG=W5_3_RNA_neuron
library(dplyr)
W5_3_RNA_neuron_RG <- SCTransform(W5_3_RNA_neuron_RG, ncells = 3000, verbose = FALSE) %>% RunPCA(verbose = FALSE) %>% 
    RunUMAP(dims = 1:30)
    
GW7_2 <- SCTransform(GW7_2, assay = "Spatial", verbose = FALSE) %>% RunPCA(verbose = FALSE)
anchors <- FindTransferAnchors(reference = W5_3_RNA_neuron_RG, query = GW7_2, normalization.method = "SCT")
predictions.assay <- TransferData(anchorset = anchors, refdata = W5_3_RNA_neuron_RG$louvain_lineage, prediction.assay = TRUE, 
    weight.reduction = GW7_2[["pca"]], dims = 1:30)
GW7_2[["predictions"]] <- predictions.assay
options(repr.plot.width=4, repr.plot.height=2)
```

```r
DefaultAssay(GW7_2) <- "predictions"
SpatialFeaturePlot(GW7_2, features = c('Myocyte-lineage','Skeletal-system-lineage','Intermediate-mesod-lineage','Cardiac-mesoderm-lineage','CNS-lineage','PNS-lineage'), pt.size.factor = 1.6, ncol = 1, crop = TRUE,combine = FALSE)
```

```r
DF=t(as.data.frame(GW7_2[["predictions"]]@data))

DF1=as.data.frame(colnames(DF)[apply(DF,1,which.max)])

colnames(DF1)='Lineage'

row.names(DF1)=row.names(DF)

GW7_2@meta.data$Lineage=as.list(DF1['Lineage'])$Lineage

GW7_2@meta.data$Lineage=as.factor(GW7_2@meta.data$Lineage)
```
